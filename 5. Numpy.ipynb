```python
first_line = [x*y for x in range(2, 100, 6) for y in range (7, 1, -2)]
second_line = [x ** 0.5 for x in range(1000, 1101, 2)]
third_line = [x**2 for x in range(51)]

big_secret = np.array([first_line, second_line, third_line, second_line, first_line])
display(big_secret)
```

```python
round(sum(list(big_secret[:,-1])),2)
```

```python
small_secret = big_secret[:,:5]
display(small_secret)
diagonal_sum = sum(small_secret[i,i] for i in range(len(small_secret)))
round(diagonal_sum,2)
```

```python
small_secret2 = big_secret[:,-5:]
display(small_secret2)
diagonal = list(small_secret2[i,i] for i in range(len(small_secret2)))

def multip(x):
    result = 1
    for i in x:
        result = i * result
    return result

multip(diagonal)
```

```python
'''Продолжим работать с массивом big_secret. Замените на 1 все элементы, 
у которых оба индекса нечётные, и на -1 все элементы, у которых оба индекса чётные.'''
big_secret[::2,::2] = -1
big_secret[1::2, 1::2] = 1
display(big_secret)

''' Выделите из каждой строки обновлённого массива big_secret первые 5 элементов. 
Чему равна сумма элементов главной диагонали получившейся матрицы? 
Введите полученный ответ без изменений и округлений.'''

new_slice = big_secret[:,:5]
new_slice_sum = sum(new_slice[i][i] for i in range(len(new_slice)))
new_slice_sum

'''Выделите из каждой строки обновлённого массива big_secret последние 5 элементов. 
Чему равно произведение элементов главной диагонали получившейся матрицы? Введите полученный результат без изменений и округлений.
'''

new_slice2 = big_secret[:,-5:] 
new_slice2_diagonal = list(new_slice2[i][i] for i in range(len(new_slice2)))
multip(new_slice2_diagonal)
```
