Write a function that receives a string as input and returns:<br>
0 - if the input received the line "Worker"<br>
1 - if the input received the line "Manager"<br>
2 - in any other case

```python
def profession_code(s):
    if s in sample['Profession'] == 'Рабочий':
        return(0)
    elif s in sample['Profession'] == 'Менеджер':
        return(1)
    else:
        return(2)
```

```python
sample2 = sample.copy()
sample2['Profession'] = sample['Profession'].apply(func=profession_code)
sample2
```

```python
def age_category(age):
    if age<23:
        return 'молодой'
    elif age>=23 and age<=35:
        return 'средний'
    else:
        return 'зрелый'
```

```python
sample['Age_category'] = sample['Age'].apply(func=age_category)
sample
```

Convert the user_id field in the log dataframe, leaving only the user ID.
In place of the entries with errors, the user_id must contain an empty string.

```python
log = pd.read_csv('data/log.xls', header=None)
log.columns = ['user_id','time','bet','win']

def clean_user_id(x):
    if '#error' in x:
        return ''
    elif 'user' in x:
        return x.replace('Запись пользователя № - ','')
log.user_id=log.user_id.apply(func=clean_user_id)
```

```python
def clean_time(x):
    if '[' in str(x):
        return x.replace('[','')
    else:
        return x

log['time'] = log['time'].apply(func=clean_time)
log
```

Read the file into the log variable;<br>
Add the names of columns user_id, time, bet, win;<br>
Remove the lines that contain user_id values with errors;<br>
Leave in the user_id field the value of the type: "user_N", where N is the identifier value;<br>
Remove the starting parenthesis from the time field.

```python
log = pd.read_csv('data/log.xls', header=None)
log.columns = ['user_id','time','bet','win']


log = log[~log.user_id.str.match('#error')]

log.user_id = log.user_id.apply(lambda s: s[s.find('user_'):])

log.time = log.time.apply(lambda s: s.replace('[',''))
log
```
