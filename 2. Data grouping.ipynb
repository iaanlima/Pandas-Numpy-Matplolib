Calculate the average and the median salaries of football players from different clubs. How many clubs have equal average and median salaries?

```python
mean_med_clubs = football.groupby(['Club'])['Wage'].agg(['mean', 'median'])
mean_med_clubs[mean_med_clubs['mean'] == mean_med_clubs['median']]\
    .sort_values(by='median', ascending=False)
```

Create a pivot table containing information on the number of players,
settling different positions in each club. Replace missing values with zeros.
What is the average number of goalkeepers (GK) in the club?

```python
pt_positions = football.pivot_table(values=['Name'],
                                    index=['Position'],
                                    columns=['Club'],
                                    aggfunc='count',
                                    # margins=True,
                                    fill_value=0
                                    )
display(pt_positions)
print(round(pt_positions.loc['GK'].mean(), 3))

# Используя таблицу, созданную на предыдущем шаге, определите,
# сколько клубов не содержат данных о центральных полузащитниках. (CM)
# Подсказка: для выполнения этого задания желательно сохранить сводную таблицу
# в виде отдельного датафрейма и сгруппировать часть данных этого
# датафрейма с помощью value_counts().
```

Create a pivot table containing the players' average speed (SprintSpeed), settled on different positions in different football clubs.<br>
Based on the data in the table, mark three positions having the highest index of average speed.

```python
df1 = football.pivot_table(values=['SprintSpeed'],
                     index=['Club'],
                     columns=['Position'],
                     margins=True,
                     fill_value=0)
print(df1.loc['All'].sort_values(ascending = False)[:3])
```

Using the table created in the previous step, mark the names of the three clubs,
in which the center forwards (ST) have the fastest average speed.

```python
df1['SprintSpeed']['ST'].sort_values(ascending=False).head(3)
```
