```python
# делаем словарь с кол-вом упоминаемых в файле ингредиентов
ingredients = set()
[[ingredients.add(y) for y in x['ingredients']] for x in recipes]

food = {} # пустой словарь для хранения информации о кол-ве ингридиентов в каждом блюде
for ingredient in ingredients:
    food[ingredient] = 0 # добавляем ингриденты в виде ключей (со значением 0)

for recipe in recipes: 
    for item in recipe['ingredients']: # перебираем значения ингредиентов в каждом рецепте
        food[item]+=1 # считаем каждый отедльный ингридиент и заносим в словарь food
```

Какие ингредиенты входят в состав более чем 20% блюд, представленных в наборе данных?
Подсказка: Напоминаем, что весь набор данных содержит сведения о 500 блюдах.

```python
# !!! dict не сортируется !!!!  
# food.items() возвращает список, поэтому мы можем его сортировать.
# (item[1] это значит что сортируется value а не key (item[0]))

dict(sorted(food.items(), key=lambda item: item[1], reverse=True))

# теперь переберём значения словаря (который принял форму списка), что бы увидеть
# какие значения встречаются по крайней мере в 20% случаев

[x[0] for x in food.items() if x[1]/500 > 0.2]
```

Сколько ингредиентов входит в состав только одного блюда?

```python
one_dish_ingred = [x[0] for x in food.items() if x[1] == 1]
len(one_dish_ingred)
```

```python
def find_item(row_cell):
    if item in row_cell:
        return 1
    return 0

for item in ingredients:
    df[item] = df['ingredients'].apply(func=find_item)
    
df['ingredients'] = df['ingredients'].apply(len)
```

Напишите код функции make_list, которая принимает строку датафрейма df, содержащую данные об одном рецепте, и возвращает перечень ингредиентов этого блюда в виде списка.

```python
# проверка
for ingredient in ingredients:         # v обащаемся к значению объекта типа series с item()
    if df[df['id'] == 22213][ingredient].item() == 1:
        print(ingredient)
```

```python
def make_list(row):
    li=[]
    for ingredient in ingredients:
        if row[ingredient].item() == 1:
            li.append(ingredient)
    return li

new_recipes = []

for current_id in ids:
    cuisine = df[df['id'] == current_id]['cuisine'].iloc[0]
    current_ingredients = make_list(df[df['id'] == current_id])
    current_recipe = {'cuisine':cuisine,'id':int(current_id),
                      'ingredients':current_ingredients}
    new_recipes.append(current_recipe)
```

```python
new_recipes
```

```python
with open('data/new_recipes.json','w') as write_file:
    json.dump(new_recipes,write_file)
```
